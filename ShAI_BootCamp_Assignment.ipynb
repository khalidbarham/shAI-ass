### Task #3 , first we should make a copy of the data so we don't mess the original one up.**bold text**

```python
cleaning = df.copy()
cleaning.head(5)
# now we drop Benefits, Notes and status columns since they are full null and they are totally useless.
cleaning = cleaning.drop(columns = ['Benefits'])
cleaning = cleaning.drop(columns = ['Notes'])
cleaning = cleaning.drop(columns = ['Status'])
cleaning.head(5)

```

```python
# After I dropped useless columns, I am now going to drop every other row containing null values since they are only a little so they won't affect my analysis
cleaning.dropna(how='any')
```

### Task #4

```python
import matplotlib.pyplot as plt

plt.hist(df['TotalPay'],bins =50)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salaries')
plt.show()
```

```python
piechart = cleaning['JobTitle'].value_counts().head(5)
plt.pie(piechart, labels=piechart, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

plt.title('Employee Distribution by Department')
```

```python
# I did not get what should be visualized in the pie chart so i am improvising :3
# First, I will make a new column names department and extract what is inside the brackets in JobTiltle column

cleaning['Department'] = cleaning['JobTitle'].str.extract(r'\((.*?)\)')

#I am going to fill department valuse that are null with others

cleaning['Department'].fillna('Others', inplace=True)

plt.pie(cleaning['Department'].value_counts().head(3),  startangle=90, colors=plt.cm.Paired.colors)
plt.title('Proportion of Employees in Different Departments')
top3_departments = cleaning['Department'].value_counts().head(3)


colors = plt.cm.Paired.colors[:len(top3_departments)]
plt.pie(top3_departments, labels=None, startangle=90, colors=colors)
for i, (label, percent, color) in enumerate(zip(top3_departments.index, top3_departments / top3_departments.sum() * 100, colors)):
    plt.text(1.2, 0.8 - i * 0.3, f"{label}: {percent:.1f}%", fontsize=10, color=color)
plt.show()
```
